```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

df = pd.read_csv('sales.csv', parse_dates=['Date'])

daily_sales = df.groupby('Date')['Sales'].sum().reset_index()
print(daily_sales)
plt.figure(figsize=(7,4))
sns.lineplot(data = daily_sales, x = 'Date', y = 'Sales',marker='o')
plt.show()
plt.savefig('sales_plot.png', dpi=300)
```

```python
categorized_sales = df.groupby('Category')['Sales'].sum().reset_index()
print(categorized_sales)
plt.figure(figsize=(5,5))
# plt.bar(categorized_sales['Category'], categorized_sales['Sales'],color='pink')
sns.barplot( x='Category', y='Sales', data=categorized_sales, palette='pastel')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Bar Chart')
plt.show()
```

```python
product_sales = df.groupby('Product')['Sales'].sum()
plt.figure(figsize=(6,6))
plt.pie(product_sales,labels=product_sales.index,autopct="%1.1f%%",startangle=140)
plt.title("Pie Chart")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.heatmap(df[['Sales','Cost','Profit']].corr(),annot=True,cmap="Greens")
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df,x="Sales",y="Profit",hue="Category",palette="Set1")
plt.title("Scatter plot")
plt.tight_layout()
plt.show()
```
